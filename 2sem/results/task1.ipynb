#TASK1: Интерполяция

```python
def interpolation(image,M):
    image_array =  image.load()
    height = image.size[0]
    width = image.size[1]
    new_height = height * M
    new_width = width * M
    result_image = pim.new('RGB',(height*M,width*M))
    result_array = result_image.load()
    for x in range(height):
      for y in range(width):
        for step_x in range(M):
          for step_y in range(M):
            result_array[x * M + step_x , y * M + step_y] = image_array[x,y]
    return result_image
    # result_image = pim.fromarray(result_array)
    # return result_image


```

#TASK2: Децимация

```python
def decimation(image,N):
  input_data = image.load()
  h0,w0 = image.size
  h1,w1 = h0//N,w0//N

  result_image = pim.new('RGB',(h1,w1))
  result_data = result_image.load()
  for x in range(h1):
    for y in range(w1):
      result_data[x,y] = input_data[x*N,y*N]
  return result_image
```

```python
decimation(input_image,100)
```

#TASK3: Передискретизация в два прохода

```python
def Oversampling_by_two(image,M,N):
  return decimation(interpolation(image,M),N)
```


#TASK4: Передискретизация в два прохода

```python
def Oversampling(image,M,N):
  input_data = image.load()
  h0,w0 = image.size
  h1,w1 = h0*M/N,w0*M/N

  result_image = pim.new('RGB',(h1,w1))
  result_data = result_image.load()

  for x in range(h1):
    for y in range(w1):
      result_data[x,y]  = input_data[x*N//M,y*N//M]
  return result_image
```
